```python
import eyed3
```

```python
file = eyed3.load("1-17. Fune.mp3")
```

```python
dir(file.tag)
```

```python
from pathlib import Path
```

```python
path = Path(file.path)
```

```python
path.name
```

```python
import re
```

```python
match = re.search(r'(\d)-(\d+)\.\s(.+)\.mp3', path.name)
```

```python
match.groups()
```

```python
file.tag.disc_num = match.group(1)
file.tag.track_num = match.group(2)
file.tag.title = match.group(3)
```

```python
file.tag.save()
```
